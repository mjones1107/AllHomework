```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import collections 
from collections import deque as deq

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#Break cities into sets
cities_grouped = [cities[x:x+50] for x in range(0, len(cities), 50)]
len(cities_grouped)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
unitsf = "imperial"


print("Beginning Data Retrieval")
print("------------------------------")
# Loop through the list of cities and perform a request for data on each
rec = 1
setplace = 1

#
cityl = []
cloud = []
country = []
date = []
humid = []
lat = []
lng = []
maxtemp = []
windspeed = []
temp = []

i = 0

while i < len(cities_grouped): #loops through city batches (13)
    #add if statement for missing cities
    for city in cities_grouped[i]: #loops through cities in the batches (50)
        parameters = {"appid" : api_key, "q" : city, "units" : unitsf}
        response = requests.get(url,params = parameters).json()
        try:
            cityl.append(response['name'])
        except:
            "id not available for " + city
        try:
            cloud.append(response['clouds']['all'])
        except:
            "cloud % not available for " + city
        try:
            country.append(response['sys']['country'])
        except:
            "country not available for " + city
        try:
            date.append(response['dt'])
        except:
            "date not available for " + city
        try:
            humid.append(response['main']['humidity'])
        except:
            "humidity % not available for " + city
        try:
            lat.append(response['coord']['lat'])
        except:
            "latitude is not available for " + city
        try:
            lng.append(response['coord']['lon'])
        except:
            "longitude is not available for " + city
        try:
            maxtemp.append(response['main']['temp_max'])
        except:
            "maximum temperature is not available for " + city
        try:
            windspeed.append(response['wind']['speed'])
        except:
            "wind speed is not available for " + city
        print("Processing Record " + str(rec) + " of Set " + str(setplace) + " | " + city)
        rec = int(rec) + 1
        response = requests.delete(url,params = parameters)
    setplace = int(setplace) + 1
    rec = 1
    i = i + 1
    time.sleep(10)
print("------------------------------")
print("Data Retrieval Complete!")
print("------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#convert to pandas data frame
weather_dataframe = pd.DataFrame({ "City" : cityl,
                              "Cloudiness": cloud, 
                              "Country" : country,
                              "Date" : date,
                              "Humidity" : humid,
                                "Lat" : lat,
                                  "Lng" : lng,
                                  "Max Temp": maxtemp,
                                  "Wind Speed": windspeed})

# Assign a "fDate" column from integer in seconds "Date" column 
weather_dataframe["fDate"] = pd.to_datetime(weather_dataframe['Date'], unit="s")
# Format the "fDate" column into a string 
weather_dataframe["fDate"] = weather_dataframe["fDate"].dt.strftime("%m/%d/%Y")

weather_dataframe.head()
```

```python
#write to csv
weather_dataframe.to_csv('output_data_file')
```
